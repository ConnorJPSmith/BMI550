Recursion Formula:  
$$
s_{i,j} = max  \left\{
\begin{array}{ll}
      s_{i-1,j-1} + 5 & x=y \\
      s_{i-1,j-1} + 1 & x\neq y \, and\, both\, are\, hydrophobic \\
      s_{i-1,j-1} -5 & x\neq y \, and\, only\, one\, is\, hydrophobic \\
      s_{i-1,j-1} + 0 & x\neq y \, and\, neither\, are\, hydrophobic \\
      s_{i-1,j} -2 & if\, opening\, gap\\
      s_{i-1,j} -1 & if\, extending\, gap\\
      s_{i,j-1} -2 & if\, opening\, gap\\
      s_{i,j-1} -1 & if\, extending\, gap\\
\end{array} 
\right. 
$$
  
This type of gap penalty punishes opening new gaps, making it so that we are more likely to continue a gap. This creates some confusion in the recursion algorithm writing, but makes our alignment less segmented. Without this gap penalty, we would also have two potential alignments with the same score (due to the adjacent W's in $s_1$)

# Question 5  
It is important to look at global and local alignment to account for various conditions. Local alignment will find better stretches of matching alignment but operates under the assumption that the rest of the alignment does not matter. Global alignment takes the entire alignment into account, but can match things incorrectly due to how to the gap penalty is set up. A researcher looking for potential mobile elements will want to use local alignment to find the mobile elements and global to verify their results by checking for alternative hypotheses.
